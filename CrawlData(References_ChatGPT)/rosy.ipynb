```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

def get_unique_image_links(page_url, exclude_list, max_links):

    driver = webdriver.Chrome()
    driver.get(page_url)

    unique_links = set()

    while len(unique_links) < max_links:

        prev_height = driver.execute_script("return document.documentElement.scrollHeight")

        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

        time.sleep(1)  


        images = driver.find_elements(By.TAG_NAME, 'img')
        for image in images:
            imgData = image.get_attribute('src')
            if imgData and imgData.startswith('https'):               
                if imgData not in unique_links:            
                    if not any(exclude_str in imgData for exclude_str in exclude_list):
                        unique_links.add(imgData)
                        if len(unique_links) == max_links:
                            break

        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        if new_height == prev_height:
            break

    driver.quit()
    for link in unique_links:
        print(link)
    return unique_links
```

```python
page_url = 'https://www.pexels.com/vi-vn/tim-kiem/hoa%20h%E1%BB%93ng/' 
exclude_list = ['logo', 'zalo', 'Web', 'facebook', 'phone', 'demo', 'lib', 'istockphoto']  
max_links = 400
IMG_LINKS=get_unique_image_links(page_url, exclude_list, max_links)
```

```python
import os
import requests

def save_links_to_directory(links, directory_path):
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)

    count = 1
    for link in links:
        filename = f"flower{count}.jpg"
        count += 1
        save_path = os.path.join(directory_path, filename)
        response = requests.get(link)
        with open(save_path, 'wb') as f:
            f.write(response.content)

        print(f" lu: {save_path}")
```
